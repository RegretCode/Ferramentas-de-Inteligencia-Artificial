# Pré-Processamento de Dados
## Tratamento dos dados
Antes de preencher valores faltantes, divida seus dados em conjuntos de treino e teste. Calcule estatísticas apenas no conjunto de treino e use-as para preencher os valores faltantes em ambos os conjuntos. Isso impede vazamento de informações do teste para o treino.

Materiais de apoio:
- https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
- https://www.freecodecamp.org/news/data-cleaning-and-preprocessing-with-pandasbdvhj/#how-to-handle-missing-values
- https://scikit-learn.org/stable/common_pitfalls.html



```python
# convertendo unidades de comprimento
def transform_data(df):
    df['Length'] = df['Length'] / 1000
    df['Width'] = df['Width'] / 1000
    df['Height'] = df['Height'] / 1000
    return df

df = transform_data(df)

# calcula massa e volume
df['Volume'] = df['Length'] * df['Width'] * df['Height'] # m³
df['Mass'] = df['Volume'] * 900 # kg
```

```python
# remove linhas com valores "not rated" ou "not tested"
df = df[~df['NCAP Rating'].isin(['Not Rated', 'Not Tested'])]
```

```python
# conjuntos de treino e teste
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)

# conversao no conjunto de treino e teste
train_df = transform_data(train_df)

train_df['Volume'] = train_df['Length'] * train_df['Width'] * train_df['Height'] # m³
train_df['Mass'] = train_df['Volume'] * 900 # kg

# remove linhas com valores "not rated" ou "not tested"
train_df = train_df[~train_df['NCAP Rating'].isin(['Not Rated', 'Not Tested'])]

# características e rótulos do conjunto de treino
X_train = train_df['Mass'].values.reshape(-1, 1)
y_train = train_df['NCAP Rating']

# transformações no conjunto de teste
test_df = transform_data(test_df)
test_df['Volume'] = test_df['Length'] * test_df['Width'] * test_df['Height'] # m³
test_df['Mass'] = test_df['Volume'] * 900 # kg
test_df = test_df[~test_df['NCAP Rating'].isin(['Not Rated', 'Not Tested'])]

# características e rótulos do conjunto de teste
X_test = test_df['Mass'].values.reshape(-1, 1)
y_test = test_df['NCAP Rating'].values.reshape(-1, 1)
```

## Treinamento dos modelos


### Árvore de Decisão
Material de apoio:
-  https://scikit-learn.org/stable/modules/tree.html

```python
# árvore de decisão
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# previsões
y_pred = clf.predict(X_test)

# avaliar o modelo
print(classification_report(y_test, y_pred, zero_division=0))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

### Máquina de Suporte de Vetores
Material de apoio:
- https://scikit-learn.org/stable/modules/svm.html

```python
# SVM
svm = SVC(kernel='linear', random_state=42)  # kernel linear para simplificar
svm.fit(X_train, y_train)

# previsões
y_pred = svm.predict(X_test)

# avaliar o modelo
print(classification_report(y_test, y_pred, zero_division=0))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

### KNN
Material de apoio:
- https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

```python
# normalizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# modelo KNN
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train_scaled, y_train)

# previsões
y_pred = knn.predict(X_test_scaled)

# avaliar o modelo
print(classification_report(y_test, y_pred, zero_division=0))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```
